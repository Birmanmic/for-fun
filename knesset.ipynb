23. flight - payment for flights to conferences
24. accommodation - payment for the MK's lodging during conferences if the conference is more than 140 kilometers from the MK's home
25. fees - judicial fees
26. travel - business trips
27. pr_literature - professional literature
28. survey - public opinion polling, provided it is related to the MP's direct activities and not a party poll
29. secretariat - payment for a secretary in the MP's personal office
30. total - the total amount spent by the MP in 2023 from the public relations budget
31. subtotal - the amount the MK could spend in 2023 (including the remaining budget from 2022)
32. norway_in - whether the KP is a new MK who joined according to the Norwegian law*
33. norway_out - MK who resigned from the Knesset according to the Norwegian law*
34. norway_kho - name of the MK who joined or left the Knesset under the Norwegian law*
35. days - the number of days the person served as an MK
36. instead_of - name of the MK who replaced another MK outside the Norwegian law*

# Let's create separate tables for regular MK and for ministers/deputy ministers, as they have different budgets and different expenditure items. 
In the future, we will analyze only regular deputies.








```python
#Creating a table for ministers and deputy ministers:



ministers_df = parliamentary_budget[(parliamentary_budget['minister'] == 'yes') | (parliamentary_budget['deputy_minister'] == 'yes')]

# Creating a table for regular deputies:
ordinary_deputies_df = parliamentary_budget[(parliamentary_budget['minister'] == 'no') & (parliamentary_budget['deputy_minister'] == 'no')]



print("\nA table for regular deputies:")
ordinary_deputies_df
```

```python

print("A table for ministers and deputy ministers:")
ministers_df
```

```python
# Add a column with daily expenses
ordinary_deputies_df['expenses_per_day'] = ordinary_deputies_df['total'] / ordinary_deputies_df['days']
ordinary_deputies_df
```

```python
# Finding the 10 most expensive and 10 least expensive deputies
top_10_expensive = ordinary_deputies_df.nlargest(10, 'expenses_per_day')
top_10_cheap = ordinary_deputies_df.nsmallest(10, 'expenses_per_day')

# Visualization
fig, ax = plt.subplots(figsize=(10, 6))

# A bar chart for the most expensive deputies
ax.bar(top_10_expensive['name'], top_10_expensive['expenses_per_day'], color='red', label='The most expensive')

# A bar chart for the least expensive
ax.bar(top_10_cheap['name'], top_10_cheap['expenses_per_day'], color='blue', label='The least expensive')

ax.set_xlabel('Deputies', fontsize=15)
ax.set_ylabel('Expenses per day NIS', fontsize=15)
ax.set_title('The most expensive and the least expensive MK', fontsize=18)
ax.legend(['Expenses'], fontsize=15)
plt.xticks(rotation=45, ha='right', fontsize=15)
plt.yticks(fontsize=15)

plt.tight_layout()
plt.show()
```

```python
# Division into coalition and opposition
coalition_df = ordinary_deputies_df[ordinary_deputies_df['coalition'] == 'yes']
opposition_df = ordinary_deputies_df[ordinary_deputies_df['coalition'] == 'no']

# Calculation of average daily expenses for coalition and opposition
avg_expenses_per_day_coalition = coalition_df['expenses_per_day'].mean()
avg_expenses_per_day_opposition = opposition_df['expenses_per_day'].mean()

# Output of results
print(f"Average daily expenses in the coalition: {avg_expenses_per_day_coalition:.2f}")
print(f"Average daily expenses in the opposition: {avg_expenses_per_day_opposition:.2f}")

# Visualization of average daily expenses
fig, ax = plt.subplots(figsize=(10, 6))

ax.bar(['Coalition', 'Opposition'], [avg_expenses_per_day_coalition, avg_expenses_per_day_opposition], color=['blue', 'red'])
ax.set_ylabel('Average daily expenses NIS', fontsize=15)
ax.set_title('Comparison of average daily expenses between coalition and opposition', fontsize=18)
ax.tick_params(axis='both', which='major', labelsize=15)

plt.show()
```

Coalition deputies spend slightly more on average than the opposition. However, it's not significant


```python
# Grouping by parties and calculating average daily expenses for each party.







party_expenses = ordinary_deputies_df.groupby('party')['expenses_per_day'].mean().reset_index()

# Renaming columns for convenience.
party_expenses.columns = ['Party', 'Average daily expenses.']

# Sorting by average daily expenses.







party_expenses = party_expenses.sort_values(by='Average daily expenses.', ascending=False)

# pruntung results
party_expenses

```

```python
print(party_expenses.columns)
```

```python
# Visualization of average daily expenses by party.


plt.figure(figsize=(10, 6))
plt.bar(party_expenses['Party'], party_expenses['Average daily expenses.'], color='green')
plt.xlabel('Party', fontsize=15)
plt.ylabel('Average daily expenses NIS', fontsize=15)
plt.title('Average daily expenses per MK by party', fontsize=18)
plt.xticks(rotation=45, fontsize=15)
plt.yticks(fontsize=15)
plt.tight_layout()
plt.show()
```

```python
# Removing columns that do not pertain to expenditure items.
expense_columns = [ 'office_needs', 'prints_publications', 'communication', 'sms', 'refreshments', 'press', 'social_networking_web', 'conference', 'cable_tv', 'male_telegrams', 'newspapers', 'mobile_phone', 'computer', 'consulting', 'office_supplies', 'language_studies', 'office_rent', 'employee_salary', 'flight', 'accomodation', 'fees', 'travel', 'pr_literture', 'survey', 'secretariat']  # добавьте остальные колонки расходов

# Grouping by parties and summing expenses for each item.
party_expenses = ordinary_deputies_df.groupby('party')[expense_columns].sum()

# To determine the most popular expense category for each party.
party_expenses['most_popular_expense'] = party_expenses.idxmax(axis=1)
party_expenses['most_popular_expense_value'] = party_expenses.max(axis=1)

# Displaying the results."
party_expenses

```

```python
print(party_expenses.columns)
```

```python
# Преобразуем индекс партий в столбец 'party'
party_expenses = party_expenses.reset_index()

# Визуализация наиболее затратных категорий расходов для каждой партии
plt.figure(figsize=(12, 8))
sns.barplot(x='party', y='most_popular_expense_value', hue='most_popular_expense', data=party_expenses)

# Настройки осей и заголовка
plt.xlabel('Party')
plt.ylabel('Total Expense  NIS')
plt.title('Most Popular Expenditure Categories by Party')
plt.xticks(rotation=45, fontsize=25)
plt.legend(title='Expense Category')

plt.tight_layout()
plt.show()
```
